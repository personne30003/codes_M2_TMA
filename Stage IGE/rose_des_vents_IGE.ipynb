```python
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from windrose import WindroseAxes
import pandas as pd
repertoire_courant=os.getcwd()
print(repertoire_courant)
repertoire_fichiers=repertoire_courant+r"\station_météo_IGE"+r"\\"
def open_file(fichier):
    Df=pd.read_csv(fichier,header=1,skiprows=lambda x: x in [2, 3],low_memory=False)
    Df['TIMESTAMP']=pd.to_datetime(Df['TIMESTAMP'].apply(lambda x:x.strip("\"")),format='mixed')
    return Df.set_index(['TIMESTAMP']).sort_index()

mois_min="05-01"
mois_max="06-30"

Df_2020=open_file(repertoire_fichiers+"2020-MTO_OSUG-B_HOUR.dat").truncate(before="2020-"+mois_min,after="2020-"+mois_max)
Df_2021=open_file(repertoire_fichiers+"2021-MTO_OSUG-B_HOUR.dat").truncate(before="2021-"+mois_min,after="2021-"+mois_max)
Df_2022=open_file(repertoire_fichiers+"2022-MTO_OSUG-B_HOUR.dat").truncate(before="2022-"+mois_min,after="2022-"+mois_max)
Df_2023=open_file(repertoire_fichiers+"2023-MTO_OSUG-B_HOUR.dat").truncate(before="2023-"+mois_min,after="2023-"+mois_max)
#Df_2024=open_file(repertoire_fichiers+"2024-MTO_OSUG-B_HOUR.dat").truncate(before="2024-"+mois_min,after="2020-"+mois_max)

print(Df_2020)
```

```python
Df_tot=pd.concat([Df_2020,
                  Df_2021,
                  Df_2022,
                  Df_2023])
fig=plt.figure()
ax = WindroseAxes.from_ax()
ax.set_title("période : "+mois_min+" "+mois_max+" toutes années confondues")
ax.bar(Df_tot['WDir_Avg'], Df_tot['WSpd_Avg'], normed=True,blowto=False, 
       opening=0.8,nsector=32, edgecolor="white")
ax.set_radii_angle()
ax.set_legend()
#plt.savefig("WindRose_IGE_mai_juin")
```

```python

#plt.title("période : 2020-"+mois_min+" "+"2020-"+mois_max)
ax = WindroseAxes.from_ax()
ax.set_title("période : 2020-"+mois_min+" "+"2020-"+mois_max)
ax.bar(Df_2020['WDir_Avg'], Df_2020['WSpd_Avg'], normed=True,blowto=False,
       opening=0.8,nsector=32, edgecolor="white",label="période : 2020-"+mois_min+" "+"2020-"+mois_max)
ax.set_radii_angle()
ax.set_legend()

#plt.title("période : 2021-"+mois_min+" "+"2021-"+mois_max)
ax = WindroseAxes.from_ax()
ax.set_title("période : 2021-"+mois_min+" "+"2021-"+mois_max)
ax.bar(Df_2021['WDir_Avg'], Df_2021['WSpd_Avg'], normed=True,blowto=False, 
       opening=0.8,nsector=32, edgecolor="white",label="période : 2021-"+mois_min+" "+"2021-"+mois_max)
ax.set_radii_angle()
ax.set_legend()

ax = WindroseAxes.from_ax()
ax.set_title("période : 2022-"+mois_min+" "+"2022-"+mois_max)
ax.bar(Df_2022['WDir_Avg'], Df_2022['WSpd_Avg'],
       normed=True,blowto=False, opening=0.8,nsector=32, edgecolor="white")
ax.set_radii_angle()
ax.set_legend()

ax = WindroseAxes.from_ax()
ax.set_title("période : 2023-"+mois_min+" "+"2023-"+mois_max)
ax.bar(Df_2023['WDir_Avg'], Df_2023['WSpd_Avg'],
       normed=True,blowto=False, opening=0.8,nsector=32, edgecolor="white")
ax.set_radii_angle()
ax.set_legend()
```
